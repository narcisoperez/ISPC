Graficar las siguientes funciones

1) a. f(x)= x

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 10)
f = x

plt.plot(x, f)
plt.title("f(x) = x")
plt.show()
```

b. f(x)=x**2

```python
x = np.linspace(-5, 5, 10)
f = x**2

plt.plot(x, f, color= 'r', label= 'x**2')
plt.legend(loc= 'upper center')
plt.title("Funcion Cuadratica")
plt.show()
```

c. f(x)= x**2 + 2

```python
x = np.linspace(-5, 5, 10)
x1 = np.linspace(-5, 5, 50)

plt.subplot(1, 2, 1)      #Sub grafica 1
f = (x**2) + 2

plt.plot(x, f, color= 'm', label= '(x**2) + 2')
plt.legend(loc= 'upper center')
plt.title("Funcion Cuadratica")

plt.subplot(1, 2, 2)      #Sub grafica 2
f = (x1**2) + 2

plt.plot(x1, f, color= 'y', label= '(x**2) + 2')

plt.xlim(-6,6)
plt.ylim(0, 10)

plt.legend(loc= 'lower center')
plt.title("Funcion Cuadratica")

plt.show()
```

d. f(x)=2**x

```python
x = np.linspace(-5, 5, 20)
f = 2**x

plt.plot(x, f, color= 'g', label= '2**x')
plt.legend(loc= 'upper center')
plt.title("Funcion Exponencial")
plt.show()
```
