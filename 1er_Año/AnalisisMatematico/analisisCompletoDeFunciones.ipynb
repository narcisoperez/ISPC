```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Dominio: 


    Si en la columna 'y' aparece el termino (inf), significa que el/los valor/es correspondiente/s de 'x' queda/n excluido del dominio.

```python
x = np.linspace(-10, 10, 21, dtype='int')
f = ((x**2) / (x-1))

tabla = pd.DataFrame(list(zip(x,f)), columns=['x', 'y'])
print("\n", tabla)
```

Asintotas:

    Asintota Vertical (A.V):
        1.Tomamos el limite, tendiendo 'x' a los N° excluidos del Dom
        2.Si el lim da inf en esos valores, hay A.V

    Asintota Horizontal (A.H):
        1.Calcular el lim de la funcion cuando x--> oo. Si existe un valor numerico, hay A.H
                                lim f(x)  = b --> y=b
                                    x--> oo
    Asintota Oblicua (A.O):
        1.Calcular pendiente(m) y ordenada al origen(n)
                     y=mx+n  --> m != 0
        2.Pendiente --> m = lim(f(x)/x)
                                x--> oo
        3.Ordenada.O --> n = lim(f(x) - mx)
                                x--> oo
        4.Las A.H y las A.O son incompatibles, si hay uno NO puede haber la otra.
        5.Una funcion racional tiene A.O cuando el grado del  numerador es una unidad mayor que el grado del denominador.
    

```python
#Asintonta Veritcal
x = sp.symbols('x')
aV = sp.limit(((x**2) / (x-1)), x, 1).doit()
print("\nEl valor de A.V es: ", aV)

#Asintota Horisontal
aH = sp.limit((x**2) / (x-1), x, 'oo').doit()
print("\nEl valor de A.H es: ", aH)

#Asintota Oblicua
m = sp.limit((x**2) / (x*(x-1)), x, 'oo').doit()                 
print("\nEl valor de M es: ", m)

n = sp.limit(((x**2 / x-1) - m*x), x, 'oo')
print("\nEl valor de N es: ", n)
```

Derivadas

```python
x = sp.symbols('x')
fx = ((x**2) / (x-1))
dx = sp.Derivative(fx, x).doit()
print("\nLa derivada de la funcion es: ", sp.simplify(dx), "\n")
```

Grafico:    

        En caso de la que funcion sea discontinua, se divide la funcion en partes y se grafica cada parte por separado.
        Caso contrario, solo se trabaja con f, x

```python
#Variables del grafico
x = np.linspace(-10, 1, num=100, endpoint=False)
x1 = np.linspace(1, 10, num=100)
x_asintota = np.linspace(-15, 15, 100)
f = (x**2) / (x-1)
f1 = (x1**2) / (x1-1)

#Funcion
plt.plot(x, f, color='b', label='Funcion')
plt.plot(x1, f1, color='b')

#Asintotas
plt.plot(x_asintota, x_asintota - 1, color='r', linestyle='--', linewidth= 1, label='A.O') #A.V

plt.plot(x_asintota*0 + 1, x_asintota * 1, color='g', linestyle='--', linewidth= 1, label='A.V')  #A.O

plt.legend(loc='lower right')
plt.title("Analisis de funciones")

plt.show()
```
